```python
from random import randint
f = open('Datas.js', "w+")
```

```python
day = 1
month = 1
for i in range(0,16):
    for j in range(0,9):
        print(j)
        if(j == 0):
            f.write("export const data"+str(i)+" = [\n")
        elif (j == 8):
            f.write("]\n\n")
        else:
            k = randint(0, 500)
            f.write("{ x: new Date('"+str(month)+"/"+str(day)+"/01'), y: "+str(k)+"},\n")
            day += 1
            if(day > 30):
                month +=1
                day = 1
                
                
for i in range(16,31):
    for j in range(0,32):
        print(j)
        if(j == 0):
            f.write("export const data"+str(i)+" = [\n")
        elif (j == 31):
            f.write("]\n\n")
        else:
            k = randint(0, 500)
            f.write("{ x: new Date('"+str(month)+"/"+str(day)+"/01'), y: "+str(k)+"},\n")
            day += 1
            if(day > 30):
                month +=1
                day = 1

f.close()
```
